```python
#B8
a,b = map(int, input().split())
if a > 0 and b > 0:
    if a > b:
        for i in range(1, a):
            if a % i == 0 and b % i == 0:
                print(i)
    else:
        for i in range(1, b):
            if a % i == 0 and b % i == 0:
                print(i)
else:
    print("Day khong phai so nguyen duong")
```
