```python
#W7A26
def add(cnt, x):
    if cnt == -1: cnt = int(x)
    else: cnt = cnt * 10 + int(x)
    return cnt

def main(a):
    numbers = []
    cnt = -1
    for i in range(len(a)):
        if (ord('0') <= ord(a[i]) <= ord('9')): cnt = add(cnt, a[i])
        else:
            if cnt != -1: numbers.append(cnt)
            cnt = -1
    if cnt != -1: numbers.append(cnt)

    total = 0
    for i in numbers: total += i
    print(total, total / len(numbers))

while 1:
    try:
        a = input()
        main(a)
    
    except EOFError:
        break
```

```python
#W7A27
def is_valid_parentheses(s):
    stack = []  # stack dùng để lưu ngoặc mở
    # dictionary ánh xạ ngoặc đóng → ngoặc mở tương ứng
    mapping = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in mapping.values():  # nếu là ngoặc mở
            stack.append(char)
        elif char in mapping:        # nếu là ngoặc đóng
            if not stack or stack[-1] != mapping[char]:
                return False         # không khớp hoặc stack rỗng → không hợp lệ
            stack.pop()               # khớp → loại bỏ ngoặc mở trên cùng
        else:
            # nếu gặp ký tự không phải ngoặc, bỏ qua (nếu muốn)
            continue

    return len(stack) == 0  # hợp lệ nếu stack rỗng

s = input("Nhập dãy ngoặc: ")
if is_valid_parentheses(s):
    print('YES')
else:
    print('NO')
```

```python
#W7A28
def update(res, l, r):
    if res[0] < r - l + 1: res = (r - l + 1, l, r)
    return res

def main(n):
    check = [1] * (n + 1)
    check[0] = check[1] = 0
    for i in range(2, n + 1):
        if check[i] == 1:
            for j in range(i * i, n + 1, i):
                check[j] = 0
    j = 4; res = (-1, 0, -1)
    for i in range(2, n):
        if check[i] == 0: continue
        else:
            res = update(res, j, i - 1)
            j = i + 1
    if check[n - 1] == 0: res = update(res, j, n - 1)
    
    for i in range(res[1], res[2]): print(i, end = ', ')
    
    print(res[2])
while 1:
    try:
        n = int(input())
        main(n)
    except EOFError:
        break
```

```python
#W7A30
def bs(a, l, r, x):
    res = r + 1
    while l <= r:
        mid = (l + r) // 2
        if a[mid] >= x:
            res = mid
            r = mid - 1
        else: l = mid + 1
    return res

def main(a):
    b = [-1] * (len(a) + 1)
    res = 0
    for i in range(len(a)):
        pos = bs(b, 0, res, a[i])
        res = max(res, pos)
        b[pos] = a[i]
    
    for i in range(1, res + 1): print(b[i], end = ' ')
    print('')

while 1:
    try:
        a = list(map(int, input().split()))
        main(a)

    except EOFError:
        break
```
