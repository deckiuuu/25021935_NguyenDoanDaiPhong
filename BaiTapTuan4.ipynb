```python
#W4A6
import math
def snt(i):
    if i < 2:
        return False
    for j in range(2, math.isqrt(i) + 1):
        if i % j == 0:
            return False
    return True
a, b = map(int, input().split())
sum = 0
x = int(x)
for x in range(a, b+1):
    if snt(x):
        sum += x
print(sum)
```

```python
#W4A14

def GCD(m, n):
    if(m < n):
        tmp = m
        m = n
        n = tmp
    if(m % n == 0):
        return n
    return GCD(n, m % n)

n, m = map(int, input().split())
print(GCD(m, n))
    
```
