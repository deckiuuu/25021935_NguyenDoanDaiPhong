```python
#W6A5
try:
    a = list(map(int, input("Nhap mang trong mot dong").split()))
except ValueError:
    print("Dau vao khong hop le, vui long nhap so nguyen")

pos = neg = ze = 0

for i in a:
    if i > 0:
        pos += 1
    elif i < 0: 
        neg += 1
    else: 
        ze += 1

x = {
    "positives" : pos,
    "negatives" : neg,
    "zeros" : ze
}

print(x)
```

```python
#W6A10
try:
    a = list(map(int, input("Nhap mang trong mot dong").split()))
    k = int(input("Nhap so nguyen k"))
except ValueError:
    print("Dau vao khong hop le, vui long nhap so nguyen")

x = []

for i in range(0, len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] != k: continue
        x.append((a[i], a[j]))

print(x)
```

```python
#W6A12
try:
    a = list(map(int, input("Nhap mang trong mot dong").split()))
except ValueError:
    print("Dau vao khong hop le, vui long nhap so nguyen")

a.sort()

dem = {}

for i in a:
    if i not in dem:
        dem[i] = 1
    else:
        dem[i] = dem.get(i) + 1

ma = 0
res = 0

for i in dem:
    x = dem.get(i)
    if ma < x:
        res = i
        ma = x

print(res)
```

```python
#W6A14
try:
    a = list(map(int, input("Nhap mang trong mot dong").split()))
    b = list(map(int, input("Nhap mang trong mot dong").split()))
except ValueError:
    print("Dau vao khong hop le, vui long nhap so nguyen")

c = []

for i in a:
    if (i in b) and (i not in c):
        c.append(i)

c.sort()
print(c)
```
