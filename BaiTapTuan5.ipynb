```python
#W5A2
a,b = map(int, input().split())
def swap(a,b):
    temp = a 
    a = b 
    b = temp
    return(a,b)
print(swap(a,b))
```

```python
#W5A3
import math
n = int(input())
def so_nguyen_to(m):
    if m < 2:
        return False
    for i in range(2, math.isqrt(m)+1):
        if n % i == 0:
            return False
    return True
if so_nguyen_to(n):
    print(n)
```

```python
#W5A4
m = int(input())
def so_hoan_hao(m):
    tong = 0
    for i in range(1, m):
        if m % i == 0:
            tong += i
            if m == tong:
                return True
    return False
if so_hoan_hao(m):
    print(m)
        
```

```python
#W5A5
a = [1,5,6,7]
k = int(input())
def kiem_tra(k):
    count = 0
    for i in (a):
        count += 1
        if i == k:
            return True
    return False
if kiem_tra(k):
    print(k)
else:
    print('-1')
```

```python
#W5A7
def toan_tu(num1, num2, kieu):
    if kieu == "+":
        ket_qua = num1+num2
    elif kieu == "-":
        ket_qua = (num1 - num2)
    elif kieu == "*":
        ket_qua = (num1*num2)
    elif kieu =="/":
        ket_qua= (num1/num2)
    return round((ket_qua),2)
num1 = float(input())
num2 = float(input())
kieu = input()
print(toan_tu(num1,num2,kieu))

    

```

```python
#W5A9
def tong2(n):
    sum = 0
    while n >0:
        sum += n%10
        n = n // 10
    return sum

n = int(input())
print(tong2(n))
```

```python
#W5A10
s1 = input()
s2 = input()
s1_to_s2 = [None] * 256
s2_to_s1 = [None] * 256
for i in range(len(s1)):
    char1 = ord(s1[i])
    char2 = ord(s2[i])
    if s1_to_s2(char1) is None:
        s1_to_s2(char1) = char2
    elif s1_to_s2[ord(s1[i])] != char2:
        print("False")
        break
    
    if s2_to_s1[ord(char2)] is None:
        s2_to_s1[ord(char2)] = char1
    elif s2_to_s1[ord(char2)] != char1:
        print("False")
        break
else:
    print("True")

```
